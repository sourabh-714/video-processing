```python
#rotating the frames 
import os
from scipy import ndimage, misc
import numpy as np

path="/Users/sourabhsharma/Desktop/ncu/internship/frames/"
rotate_img="/Users/sourabhsharma/Desktop/ncu/internship/rotate_180/"
count=0
for image in os.listdir(path):
    input=os.path.join(path,image)
    target_image=(ndimage.imread(input))
    
    rotate=ndimage.rotate(target_image,180)
    
    out_path=os.path.join(rotate_img,"rotated%d.jpg"%count)
    misc.imsave(out_path,rotate)
    count+=1

   
    
```
